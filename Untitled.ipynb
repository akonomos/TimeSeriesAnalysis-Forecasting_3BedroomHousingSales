```python
#select houses with 3 bedrooms from dataset 
subset_df = df2[(df2['type'] == 'house') & (df2['bedrooms'] == 3)]
subset_df
```

```python
#clean df to fit prophet
#input to Prophet is always a dataframe with two columns: ds and y
subset_df = subset_df.drop(labels='type', axis=1)
```

```python
subset_df
```

```python
subset_df.isnull().sum()
```

```python
subset_df.info()
```

```python
#convert saledate to prophet format by renaming columns
subset_df.rename(columns={"y": "ds"}, inplace = True)
subset_df.rename(columns={"MA": "y"}, inplace = True)
subset_df
```

```python
#fit the model by instantiating a new Prophet 
m = Prophet(interval_width = 0.90, daily_seasonality = True)
m.fit(subset_df)
```

```python
#predictions for the future
future_df2 = m.make_future_dataframe(periods = 200, freq ="D")
forecast = m.predict(future_df2)
forecast.head(3)
```

```python
future_df2
```

```python
forecast[["ds", "yhat","yhat_upper", "yhat_lower"]].head(3)
```

```python
m.plot(forecast, include_legend = True)
plt.title('Forecast for the next 200 days for 3 Bedroom Houses')
plt.show()
#this forecast is faulty because our prices were measured averages- check out df1 
```

```python
#mean absolute error from expected and predicted values
from sklearn.metrics import mean_absolute_error
y_true = subset_df['y'].values
y_pred = forecast['yhat'][:-200].values
mae = mean_absolute_error(y_true, y_pred)
print('Mean Absolute Error: %.3f' % mae)
print('This means there is an average error of 994.543 between our actual value and the predicted value')
```

```python
plt.figure(figsize=(16,6))
plt.plot(y_true, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.title('Actual and Predicted Curve of the Sales for 3 Bedroom Houses')
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.style.use('seaborn')
plot  = m.plot_components(forecast)
plt.title('Traffic of Sales')
plt.show()
```

```python
#load data
df1 = df1[(df1['propertyType'] == 'house') & (df1['bedrooms'] == 3)]
df1
```

```python
#clean df to fit prophet
#input to Prophet is always a dataframe with two columns: ds and y
df1 = df1.drop(labels='postcode', axis=1)
df1 = df1.drop(labels='bedrooms', axis=1)
df1 = df1.drop(labels='propertyType', axis=1)
```

```python
df1
```

```python
#convert saledate to prophet format by renaming columns
df1.rename(columns={"datesold": "ds"}, inplace = True)
df1.rename(columns={"price": "y"}, inplace = True)
df1
```

```python
#fit the model by instantiating a new Prophet 
m1 = Prophet(interval_width = 0.90, daily_seasonality = True)
m1.fit(df1)
```

```python
#predictions for the future
future_df1 = m1.make_future_dataframe(periods = 200, freq ="D")
forecast = m1.predict(future_df1)
forecast.head(3)
```

```python
m1.plot(forecast, include_legend = True)
plt.title('Forecast for the next 200 days for 3 Bedroom Houses')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
y_true = df1['y'].values
y_pred = forecast['yhat'][:-200].values
plt.figure(figsize=(16,6))
plt.plot(y_true, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.title('Actual and Predicted Curve of the Sales for 3 Bedroom Houses')
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.style.use('seaborn')
plot  = m1.plot_components(forecast)
plt.title('Traffic of Sales')
plt.show()
```
